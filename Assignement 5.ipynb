```python
#9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the
#fraction as "numerator/denominator".
class Fraction:
    def __init__(self, numerator, denominator):
        """
        Initializes a Fraction object.

        Args:
            numerator (int): The numerator of the fraction.
            denominator (int): The denominator of the fraction.
        """
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        """
        Returns a string representation of the fraction in "numerator/denominator" format.
        """
        return f"{self.numerator}/{self.denominator}"

# Demonstrate the Fraction class
if __name__ == "__main__":
    fraction1 = Fraction(3, 4)
    fraction2 = Fraction(7, 2)
    fraction3 = Fraction(-5, 8)

    print(f"Fraction 1: {fraction1}")
    print(f"Fraction 2: {fraction2}")
    print(f"Fraction 3: {fraction3}")

    # Example of handling a zero denominator
    try:
        invalid_fraction = Fraction(1, 0)
        print(invalid_fraction)
    except ValueError as e:
        print(f"Error creating fraction: {e}")




```

```python
#10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
#vectors.
class Vector:
     def __init__(self, x, y):
        self.x = x
        self.y = y
     def __str__(self):
        return f"Vector({self.x}, {self.y})"

     def __add__(self, other):
        if isinstance(other, Vector):
            new_x = self.x + other.x
            new_y = self.y + other.y
            return Vector(new_x, new_y)
        else:
            raise TypeError("Unsupported operand type for +: 'Vector' and '{}'".format(type(other).__name__))

# Create Vector objects
v1 = Vector(1, 2)
v2 = Vector(3, 4)

# Add two vectors using the overloaded + operator
v3 = v1 + v2

# Print the resulting vector
print(v3)

# Demonstrate adding a Vector to a non-Vector type (will raise an error)
try:
    v4 = v1 + 5
except TypeError as e:
    print(e)



```

```python
#12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
#the average of the grades.
class Student:
    def __init__(self, name, grades):
        """
        Initializes a Student object with a name and a list of grades.

        Args:
            name (str): The name of the student.
            grades (list): A list of numerical grades for the student.
        """
        self.name = name
        self.grades = grades

    def average_grade(self):
        """
        Computes and returns the average of the student's grades.

        Returns:
            float: The average grade, or 0.0 if there are no grades.
        """
        if not self.grades:
            return 0.0  # Return 0 if there are no grades to avoid division by zero
        return sum(self.grades) / len(self.grades)

# Example usage:
student1 = Student("Alice", [90, 85, 92, 88])
print(f"{student1.name}'s average grade: {student1.average_grade()}")

student2 = Student("Bob", [75, 80, 78])
print(f"{student2.name}'s average grade: {student2.average_grade()}")

student3 = Student("Charlie", [])
print(f"{student3.name}'s average grade: {student3.average_grade()}")


```

```python
#14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
#and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
    def __init__(self, name, hourly_rate):
        self.name = name
        self.hourly_rate = hourly_rate

    def calculate_salary(self, hours_worked):
        """
        Calculates the salary based on hours worked and hourly rate.
        """
        return hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hourly_rate, bonus):
        super().__init__(name, hourly_rate)  # Call the parent class's constructor
        self.bonus = bonus

    def calculate_salary(self, hours_worked):
        """
        Calculates the salary, adding a bonus to the base employee salary.
        """
        base_salary = super().calculate_salary(hours_worked)  # Get salary from parent
        return base_salary + self.bonus

# Example Usage:
if __name__ == "__main__":
    # Create an Employee object
    employee1 = Employee("Alice", 25)
    employee_salary = employee1.calculate_salary(160)
    print(f"{employee1.name}'s salary: ${employee_salary:.2f}")

    # Create a Manager object
    manager1 = Manager("Bob", 35, 1000)
    manager_salary = manager1.calculate_salary(160)
    print(f"{manager1.name}'s salary: ${manager_salary:.2f}")
```

```python
#Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.
class House:
    """A class to represent a generic house."""

    def __init__(self, address, price):
        """Initializes the House object."""
        self.address = address
        self.price = price

    def get_info(self):
        """Returns the basic information about the house."""
        return f"Address: {self.address}, Price: ${self.price:,.2f}"

class Mansion(House):
    """A derived class representing a mansion, which is a type of house."""

    def __init__(self, address, price, number_of_rooms):
        """Initializes the Mansion object, inheriting from House."""
        super().__init__(address, price)  # Call the parent class constructor
        self.number_of_rooms = number_of_rooms

    def get_info(self):
        """Overrides the parent method to include room details."""
        house_info = super().get_info()  # Get base information from the parent
        return f"{house_info}, Rooms: {self.number_of_rooms}"

# --- Demonstrate the classes ---
if __name__ == "__main__":
    # Create an instance of the base class (House)
    my_house = House("123 Main St", 350000)
    print("House Information:")
    print(my_house.get_info())

    print("-" * 25)

    # Create an instance of the derived class (Mansion)
    my_mansion = Mansion("456 Grand Ave", 2500000, 20)
    print("Mansion Information:")
    print(my_mansion.get_info())

```

```python
""""

Question and answer
1) What is Object-Oriented Programming (OOP)?

What are four basic principles of Object Oriented Programming ...
Object-Oriented Programming (OOP) is a programming paradigm that models software as a collection of interacting objects, 
which are instances of classes (blueprints) containing both data (attributes) and methods (functions) that operate on that data. 
It organizes code around objects, promoting modularity, reusability, and easier maintenance by bundling data and behavior together 
through key principles like encapsulation, inheritance, polymorphism, and abstraction.  

2) What is a class in OOP?
In Object-Oriented Programming (OOP), a class serves as a blueprint or a template for creating objects. It defines the structure and behavior that objects of that class will possess. 
Here's a breakdown of what a class encapsulates:

    Attributes (Data Members/Properties):
    These are the characteristics or data associated with an object. For example, in a Car class, attributes might include model, color, and speed.
    Methods (Member Functions/Behaviors):
    These are the functions or operations that objects of the class can perform. In the Car class example, methods could be accelerate(), brake(), or startEngine(). 


3) What is an object in OOP?

In Object-Oriented Programming (OOP), an object is a fundamental concept representing a real-world entity or a conceptual component within a program. 
It is an instance of a class, which serves as a blueprint defining the object's structure and behavior.
Here's a breakdown of what constitutes an object in OOP:

    Instance of a Class:
    While a class is a template or definition, an object is a concrete realization of that class. When you create an object, you are instantiating the class, and memory is allocated for that specific object.
    Encapsulates Data and Functionality:
    Objects combine both data (attributes or properties) and methods (behaviors or functions) that operate on that data. This encapsulation keeps related data and functionality together, promoting modularity and data integrity.
    State:
    The state of an object is represented by the values of its attributes at a particular point in time. For example, a "Car" object's state might include its color, speed, and current gear.
    Behavior:
    The behavior of an object is defined by its methods, which dictate how the object can interact with other objects or respond to external stimuli. For example, a "Car" object might have methods like start(), accelerate(), or brake().
    Identity:
    Each object typically possesses a unique identity, allowing it to be distinguished from other objects, even if they are instances of the same class and have identical states. 

In essence, objects are the active participants in an OOP program, interacting with each other to achieve the desired functionality. They provide a structured way to model and manage complexity by breaking down a system into self-contained, interacting units.

4) What is the difference between abstraction and encapsulation?
 Abstraction focuses on hiding implementation details and exposing only essential features, defining what an object does, while encapsulation bundles data and methods together into a single unit, controlling access and protecting the internal state, thus ensuring how the object's data is managed and secured. Abstraction is a design-level concept often implemented with interfaces and abstract classes, whereas encapsulation is an implementation-level mechanism that uses access modifiers to hide code and data. 
Abstraction

    Focus: Hiding complexity and showing only the necessary information. 

What it does: Focuses on defining what an object does, not how it does it. 
Achieved through: Abstract classes and interfaces, which define a contract for behavior without providing the implementation. 
Goal: To reduce complexity and make systems easier to use by simplifying the representation of objects. 
Analogy: A car's steering wheel and pedals abstract the complex mechanics of driving; you know what they do, but not how the engine works. 

Encapsulation

    Focus:
    Protecting the internal data and methods of an object. 

What it does:
Bundles data and the methods that operate on that data into a single unit, like a class. 
Achieved through:
Access modifiers (like public and private) to control visibility and access to the class's members. 
Goal:
To ensure data integrity and privacy by restricting direct access and providing controlled access through methods (getters and setters). 
Analogy:
The internal circuits and components of a car engine are encapsulated, hidden under the
hood, and only accessible through controlled interfaces. 

5) What are dunder methods in Python?

Dunder methods, also known as magic methods or special methods, are a core feature of Python's object model. 
The term "dunder" is a contraction of "double underscore," referring to the two underscores that prefix and suffix their names (e.g., __init__, __add__, __len__).

6) Explain the concept of inheritance in OOP?
A subclass automatically gains access to the public and protected attributes and methods of its superclass.
The subclass can then add its own unique attributes and methods, or it can override (redefine) methods inherited from the superclass to provide specialized functionality.

7) What is polymorphism in OOP?

Polymorphism in OOP is the concept that an object, method, or variable can take on many forms. It allows a 
single interface to represent different underlying forms or behaviors, making programs more flexible and adaptable
 by enabling you to handle objects of different types uniformly through a common interface. Key types include compile-time (static) polymorphism, 
 achieved through method overloading, and runtime (dynamic) polymorphism, implemented via method overriding.  

8) How is encapsulation achieved in Python?

Encapsulation in Python, while not as strictly enforced as in some other object-oriented languages, is primarily achieved through a combination of conventions and features:

    Classes:
    The fundamental unit of encapsulation in Python is the class. Classes bundle data (attributes) and the methods that operate on that data into a single, cohesive unit.
    Access Modifiers (by convention):
    Python does not have explicit public, private, or protected keywords. Instead, it uses naming conventions to indicate intended access levels: 

9) What is a constructor in Python?
 In Python, a constructor is a special method used to initialize objects when they are created from a class. Its primary purpose is to set up the initial state of the object by assigning values to its attributes. 
Here are the key points about constructors in Python:

    __init__ Method:
    In Python, the constructor is defined using the special method __init__. This method is automatically called whenever a new object (instance) of a class is created.
    Initialization:
    The __init__ method allows you to define what happens when an object is instantiated. You typically use it to assign initial values to the object's attributes, ensuring that the object starts in a consistent and usable state.
    self Parameter:
    The first parameter of the __init__ method (and all instance methods in Python) is conventionally named self. This parameter refers to the instance of the object itself and is used to access and modify its attributes.
    Parameters:
    The __init__ method can accept additional parameters, allowing you to pass initial data to the object during its creation. These parameters are then used to set the values of the object's attributes
10) What are class and static methods in Python?

In Python, class methods and static methods are special types of methods within a class that differ from regular instance methods in how they interact with the class and its instances.
1. Class Methods:

    Definition: A class method is bound to the class and receives the class itself as its first argument (conventionally named cls). 
    It is defined using the @classmethod decorator.
    Purpose: Class methods are primarily used for operations that involve the class's state or for creating alternative constructors.
      They can access and modify class-level attributes but cannot directly access or modify instance-specific attributes
11) What is method overloading in Python?
 Method overloading in Python refers to the concept of having multiple methods with the same name within a single class, but with different parameters or arguments. While Python does not support method overloading in the same way as statically-typed languages like Java or C++, it provides mechanisms to achieve similar functionality. 
Achieving Method Overloading in Python:

    Default Arguments: You can define a single method with default values for its parameters, allowing it to be called with varying numbers of arguments. 


12) What is method overriding in OOP?
 Method overriding in OOP allows a subclass to provide a specific implementation for a method that is already defined in its parent class,
   a process that is also known as runtime polymorphism. Key requirements include inheritance>> (the subclass must inherit from the parent) and
     a matching method signature>> (same name, return type, and parameters). This enables the subclass to customize inherited behavior, with the version of 
     the method executed depending on the actual object type at runtime. 

13 What is a property decorator in Python?
 The @property decorator in Python is a built-in feature that allows methods within a class to be accessed and managed like attributes. 
 It provides a way to encapsulate attribute access, enabling the definition of custom logic for getting, setting, and deleting an attribute's value 
 without explicitly calling getter and setter methods. 

14) Why is polymorphism important in OOP?
 Polymorphism is important in OOP because it enables code reusability, flexibility, and extensibility by allowing objects of different types to be 
 treated as objects of a common supertype, respond to the same method calls in their own specific ways, and for new subclasses to be added without altering existing code. 
 It is a fundamental concept that distinguishes object-oriented languages from object-based ones, making systems more adaptable and maintainable.  

15) What is an abstract class in Python?
 In Python, an abstract class is a class that cannot be instantiated directly and is designed to serve as a blueprint or template for other classes. 
 It defines a common interface that its subclasses must adhere to, ensuring a consistent structure and behavior across related classes. 

16) What are the advantages of OOP?
 The primary advantages of Object-Oriented Programming (OOP) are code reusability, making programs faster and more efficient, and modularity, 
 which simplifies debugging and maintenance by breaking down complex systems into smaller, self-contained objects. OOP also enhances extensibility and flexibility 
 through features like polymorphism and inheritance, allowing for adaptable software that can be easily updated. Other benefits include improved code organization, 
 enhanced security, and increased developer productivity. 

17) What is the difference between a class variable and an instance variable?

The fundamental difference between a class variable and an instance variable lies in their scope, ownership, and how they are stored and accessed.
Class Variable:

    Ownership:
    Belongs to the class itself, not to any specific instance of the class.
    Storage:
    Only one copy of a class variable exists in memory, regardless of how many objects (instances) of the class are created. All instances share this single copy.
    Declaration:
    Typically declared directly within the class definition, outside of any methods. In some languages like Java, they are often declared using the static keyword.
    Access:
    Can be accessed using the class name (e.g., ClassName.variable_name) or through an instance (e.g., instance_name.variable_name), though accessing through the class name is generally preferred for clarity.
    Use Cases:
    Suitable for data that is common to all instances of a class, such as constants, counters, or shared configurations. 

Instance Variable:

    Ownership:
    Belongs to a specific instance (object) of the class.
    Storage:
    Each object created from the class gets its own independent copy of the instance variables. Changes to an instance variable in one object do not affect the same variable in other objects. 

Declaration:
Typically declared within the constructor or other methods of the class, often by assigning values to self.variable_name (in Python) or simply variable_name (in Java, within a class but outside a method).
Access:
Can only be accessed through an object (instance) of the class (e.g., instance_name.variable_name).
Use Cases:
Suitable for data that represents the unique state or characteristics of an individual object.
18) What is multiple inheritance in Python?

Multiple inheritance in Python is a feature that allows a class to inherit attributes and methods from more than one parent class. This means a 
single child class can combine functionalities from multiple distinct sources.

19) Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python?

In Python, __str__ and __repr__ are special "dunder" methods used to define how an object is represented as a string. They serve different purposes and are aimed at different audiences:
1. __str__ (for "string"):

    Purpose:
    To provide a human-readable, user-friendly string representation of an object. This is what an end-user would typically want to see when an object is printed or converted to a string.
    Usage:
    It's invoked implicitly when you use print() on an object, str() to convert an object to a string, or when an object is formatted within f-strings.
    Characteristics:
    The output should be concise, clear, and easily understandable, focusing on the most relevant information for a user.
20) What is the significance of the ‘super()’ function in Python?

The super() function in Python holds significant importance within the context of object-oriented programming, particularly concerning inheritance. Its primary role is to provide a mechanism for accessing methods and attributes of a parent or sibling class from within a child class.
Here's a breakdown of its significance:

Method Overriding and Extension: When a child class overrides a method from its parent, super() allows the child to still call and utilize the parent's implementation of that method. This enables code reuse and allows for extending the parent's functionality rather than completely replacing it. 


21) What is the significance of the __del__ method in Python?

The __del__ method in Python, often referred to as a destructor, holds significance primarily in resource management and cleanup operations when an object is about to be destroyed.
Here's a breakdown of its significance:

    Destructor Functionality:
    It provides a mechanism to define actions to be executed just before an object is garbage collected and its memory reclaimed. This contrasts with the __init__ method (constructor), which handles object initialization.
    Resource Release:
    Its main purpose is to release external resources held by the object that are not automatically managed by Python's garbage collector. Examples include:
        Closing file handles.
        Terminating network connections.
        Releasing database connections.
        Freeing up memory allocated by external libraries. 
    Automatic Invocation:
    The __del__ method is automatically called by Python's garbage collector when an object's reference count drops to zero, meaning there are no longer any references to that object in the program. You do not explicitly call __del__ yourself
22) What is the difference between @staticmethod and @classmethod in Python?

The primary distinction between @staticmethod and @classmethod in Python lies in their access to the class and instance.
@classmethod:

    Accesses the class:
    A class method receives the class itself as its first argument, conventionally named cls. This allows the method to access and potentially modify class-level attributes.
    Use cases:
    Commonly used for creating alternative constructors (factory methods) or methods that operate on class-level data, such as modifying a class attribute that applies to all instances. 

@staticmethod:

    No access to class or instance:
    A static method does not receive self (instance) or cls (class) as an argument. It behaves like a regular function defined within the class's namespace.
    Use cases:
    Primarily used for utility functions that are logically related to the class but do not depend on the state of an instance or the class itself. They do not interact with class or instance attributes. 


23) How does polymorphism work in Python with inheritance?

Polymorphism in Python, particularly in the context of inheritance, functions primarily through method overriding. This allows objects of different classes, all derived from a common base class, to respond to the same method call in a way specific to their own class.
Here's how it works:

    Inheritance:
    A child class (subclass) inherits methods and attributes from a parent class (superclass). This establishes a "is-a" relationship, where a child class is a type of the parent class. 

Method Overriding:
A child class can provide its own implementation for a method that is already defined in its parent class. When an instance of the child class calls that method, its own specific implementation is executed instead of the parent's. 
Polymorphic Behavior:
Because of this overriding, you can treat objects of different subclasses uniformly through a common interface (the parent class). A function or piece of code designed to work with the parent class type can accept instances of any of its subclasses, and the correct, overridden method will be called automatically based on the object's actual type. 

24) What is method chaining in Python OOP?
 Method chaining in Python OOP is a programming technique that allows for calling multiple methods on an object in a single, continuous line of code. This is achieved by having each method return the object itself (typically self) after performing its operation. 
How it works:

    Return self: In a class, each method designed for chaining must explicitly return self at the end of its execution. This ensures that after a method is called, the object instance is returned, allowing the next method in the chain to be called on the same object

25) What is the purpose of the __call__ method in Python?
 The __call__ method in Python serves the purpose of making instances of a class callable, meaning they can be invoked like functions. When you define __call__ 
 within a class, you can then "call" an object of that class using parentheses, just as you would call a regular function. 

"""
```
