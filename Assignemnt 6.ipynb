```python
#4.Write a Python script that reads from one file and writes its content to another file
import os

def copy_file_content(source_filepath, destination_filepath):
    print(f"Attempting to copy content from '{source_filepath}' to '{destination_filepath}'...")
    
    try:
        with open(source_filepath, 'r', encoding='utf-8') as source_file:
            content = source_file.read()
            print(f"Successfully read {len(content)} characters from the source file.")
            
        with open(destination_filepath, 'w', encoding='utf-8') as destination_file:
            destination_file.write(content)
            print("Copy completed successfully!")
            
    except FileNotFoundError:
        print(f"\nError: The source file '{source_filepath}' was not found.")
        print("Please ensure 'source.txt' exists in the same directory as the script.")
    except Exception as e:
        print(f"\nAn unexpected error occurred during file operation: {e}")


SOURCE_FILE = "source.txt"
DESTINATION_FILE = "destination.txt"
copy_file_content(SOURCE_FILE, DESTINATION_FILE)

print(f"\nCheck the contents of '{DESTINATION_FILE}' to see the copied text.")


```

```python
# Write a Python program that uses a try-except block to handle an error when attempting to access adictionary key that doesn't exist
def safe_dictionary_access(dictionary, key):
    
    try:
        value = dictionary[key]
        print(f"Successfully retrieved value: '{value}' for key: '{key}'")
        return value
    except KeyError:
        print(f"Error: The key '{key}' does not exist in the dictionary.")
        print("Accessing a non-existent key raises a KeyError.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

data = {
    "product_id": "P001",
    "name": "Laptop",
    "price": 1200.00
}

print("--- Test 1: Existing Key ---")
safe_dictionary_access(data, "name")
print("-" * 30)

print("--- Test 2: Non-Existent Key ---")
safe_dictionary_access(data, "stock")
print("-" * 30)

print("--- Test 3: Another Non-Existent Key ---")
safe_dictionary_access(data, "quantity")
print("-" * 30)
```

```python
#Write a program that demonstrates using multiple except blocks to handle different types of exceptions
def robust_calculator():
    
    print("Welcome to the Robust Calculator! Let's divide two numbers.")

    try:
        num1_str = input("Please enter the numerator (the number on top): ")
        num2_str = input("Please enter the denominator (the number on the bottom): ")

        numerator = float(num1_str)
        denominator = float(num2_str)

        
        result = numerator / denominator

        print("\n--- Success! ---")
        print(f"The result of {numerator} divided by {denominator} is: {result}")

    except ValueError:
        print("\n--- Input Error! ---")
        print("I'm so sorry, but you must enter valid numbers (e.g., 5, 10.5).")
        print("I caught a **ValueError** because I couldn't convert your text into a number.")

    except ZeroDivisionError:
        print("\n--- Math Error! ---")
        print("Oops! You cannot divide a number by zero.")
        print("I caught a **ZeroDivisionError** and prevented the program from crashing!")

    except Exception as e:
        print("\n--- Unexpected Error! ---")
        print(f"An unexpected problem occurred: {e}")
        print("This is a catch-all block for errors we didn't specifically anticipate.")

    finally:
        print("\nThank you for using the Robust Calculator!")


```

```python
# How would you implement a basic logging setup that logs to a file with rotation after 1MB

import logging
import logging.handlers
import os

# --- Configuration Constants ---
LOG_FILENAME = "app_activity.log"
MAX_LOG_SIZE_BYTES = 1024 * 1024  # 1 MB (1024 KB/MB * 1024 Bytes/KB)
BACKUP_COUNT = 3  # Keep 3 old log files (e.g., app_activity.log.1, .2, .3)
LOG_LEVEL = logging.INFO
LOG_FORMAT = '%(asctime)s - %(levelname)s - %(message)s'

def setup_rotating_logger():
    """
    Sets up a logger that logs messages to a file with rotation.
    """
    # 1. Get the root logger
    logger = logging.getLogger()
    logger.setLevel(LOG_LEVEL)

    # 2. Create a RotatingFileHandler
    # This is the core component that manages file size and rotation.
    handler = logging.handlers.RotatingFileHandler(
        filename=LOG_FILENAME,
        maxBytes=MAX_LOG_SIZE_BYTES,
        backupCount=BACKUP_COUNT,
        encoding='utf-8'
    )

    # 3. Create a formatter and set it for the handler
    formatter = logging.Formatter(LOG_FORMAT)
    handler.setFormatter(formatter)

    # 4. Add the handler to the logger
    logger.addHandler(handler)
    
    # Optional: Add a StreamHandler to see output on the console as well
    console_handler = logging.StreamHandler()
    console_handler.setFormatter(formatter)
    logger.addHandler(console_handler)

    print(f"Logger '{LOG_FILENAME}' configured with rotation at {MAX_LOG_SIZE_BYTES / (1024*1024):.0f}MB.")
    return logger

# --- Program Execution ---
if __name__ == "__main__":
    
    # Clean up old logs for a fresh run (optional)
    if os.path.exists(LOG_FILENAME):
        os.remove(LOG_FILENAME)
        print(f"Removed old log file: {LOG_FILENAME}")

    # Set up the logger
    app_logger = setup_rotating_logger()

    # Log some messages
    app_logger.info("Application started successfully.")
    app_logger.warning("Configuration file not found, using defaults.")
    
    # To demonstrate rotation, you would typically write a large number of logs
    # or reduce the MAX_LOG_SIZE_BYTES to a very small number (e.g., 100 bytes).
    # We will use a small size here to ensure the rotation demonstration works instantly.
    
    app_logger.info("Demonstrating file rotation (by setting log size to a tiny value)...")
    
    # Temporarily reset the handler for a small file size demo (for immediate rotation)
    tiny_size_bytes = 100
    demo_handler = logging.handlers.RotatingFileHandler(
        filename=LOG_FILENAME,
        maxBytes=tiny_size_bytes,
        backupCount=3,
        encoding='utf-8'
    )
    demo_handler.setFormatter(logging.Formatter(LOG_FORMAT))
    # Remove the old handler and add the new one for the demo
    for h in app_logger.handlers[:]:
        app_logger.removeHandler(h)
    app_logger.addHandler(demo_handler)
    
    
    # Write messages that will force a rotation
    for i in range(10):
        app_logger.info(f"Log message number {i:02d}. This message is written to force a file rotation.")

    app_logger.info("Finished logging messages. Check the directory for rotated files.")
```

```python
#Write a program that handles both IndexError and KeyError using a try-except block



def access_data_safely(data_structure, access_key):
    """
    Attempts to access an element in either a list/tuple or a dictionary.
    Handles IndexError (for sequences) and KeyError (for dictionaries).

    :param data_structure: A list, tuple, or dictionary.
    :param access_key: The index (int) or key (string) to use for access.
    """
    print(f"\n--- Attempting to access '{access_key}' in {type(data_structure).__name__} ---")
    
    try:
        # The access attempt is placed in the 'try' block.
        # This single line is capable of raising BOTH KeyError (if a dict) 
        # or IndexError (if a list/tuple).
        value = data_structure[access_key]
        
        # If successful, print the result
        print(f"Success! The value retrieved is: '{value}'")

    except KeyError:
        # --- Specific Handler for Dictionaries ---
        print("\n*** ERROR: KeyError ***")
        print(f"You tried to access a dictionary key ('{access_key}') that does not exist.")
        print("Please check the available keys and try again.")
        
    except IndexError:
        # --- Specific Handler for Lists/Tuples ---
        print("\n*** ERROR: IndexError ***")
        print(f"You tried to access an index ('{access_key}') that is out of range.")
        print("Remember, indices start at 0, and the list/tuple isn't that long!")
        
    except Exception as e:
        # --- General Handler for anything else (optional, but good practice) ---
        print("\n*** UNEXPECTED ERROR ***")
        print(f"An unexpected error occurred: {e}")

# ======================================================
# 1. Dictionary Tests (KeyError scenarios)
# ======================================================
person_dict = {"name": "Alice", "age": 30}

# Test 1a: Successful access
access_data_safely(person_dict, "name")

# Test 1b: Accessing a non-existent key (raises KeyError)
access_data_safely(person_dict, "city")

# ======================================================
# 2. List Tests (IndexError scenarios)
# ======================================================
colors_list = ["Red", "Green", "Blue"]

# Test 2a: Successful access
access_data_safely(colors_list, 1) # Accesses 'Green'

# Test 2b: Accessing an index out of bounds (raises IndexError)
access_data_safely(colors_list, 5)

# ======================================================
# 3. Mismatched Type Test (Raises a different error, caught by general handler)
# ======================================================
# Note: Accessing a list with a string key (raises TypeError)
access_data_safely(colors_list, "key")
```

```python
#How would you open a file and read its contents using a context manager in Python
def read_file_with_context_manager(filename):
    
    try:
        # The 'with' statement is the context manager.
        # open(filename, 'r') opens the file for reading.
        # 'as file' assigns the file object to the variable 'file'.
        with open(filename, 'r') as file:
            print(f"File '{filename}' opened successfully.")
            
            # Use the .read() method to get the entire content as a single string
            content = file.read()
            
            # Print the content
            print("\n--- Content Read ---")
            print(content.strip())
            print("--------------------\n")
            
        # The file is automatically closed here when the 'with' block finishes.
        print(f"File '{filename}' is now automatically closed.")
        
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# --- Demonstration Setup ---
# Create a temporary file to read from
temp_filename = "data_to_read.txt"
try:
    with open(temp_filename, 'w') as f:
        f.write("Welcome to the safest way to read files!\n")
        f.write("The context manager handles all cleanup.\n")
except IOError:
    # If the setup fails, we can't run the main demo
    print("Could not create the sample file for reading.")

# --- Execution ---
read_file_with_context_manager(temp_filename)
```

```python
# Write a Python program that reads a file and prints the number of occurrences of a specific word
import string

def count_word_occurrences(filename, target_word):
    count = 0
    
    # Normalize the target word to lowercase for case-insensitive comparison
    target_word_lower = target_word.lower()

    try:
        # Use a context manager (with open...) for safe file handling
        with open(filename, 'r', encoding='utf-8') as file:
            # Read the entire file content at once
            content = file.read()
            
            # --- Text Preprocessing ---
            
            # 1. Convert all text to lowercase
            content_lower = content.lower()
            
            # 2. Remove punctuation (optional, but makes counting more accurate)
            # We use str.maketrans and str.translate for efficient removal
            translator = str.maketrans('', '', string.punctuation)
            content_cleaned = content_lower.translate(translator)
            
            # 3. Split the cleaned text into a list of words
            word_list = content_cleaned.split()
            
            # --- Counting ---
            
            # Iterate through the list and count matches
            for word in word_list:
                if word == target_word_lower:
                    count += 1

    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found. Please check the file path.")
        return -1 # Return -1 to indicate an error
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return -1

    return count

# --- Demonstration Setup ---
# 1. Define the filename and the word to search for
file_to_check = "sample_document.txt"
search_term = "python"

# 2. Create a sample file for demonstration
try:
    with open(file_to_check, 'w', encoding='utf-8') as f:
        f.write("Python is a powerful programming language. "
                "I love Python! This sentence is about the Python language.\n"
                "You should learn Python today. PYTHON is fun.")
except IOError:
    print("Could not create the sample file. Aborting demonstration.")
    exit()

# --- Execution ---
print(f"Searching for the word: '{search_term}'")
print("-" * 40)

# Call the function
occurrences = count_word_occurrences(file_to_check, search_term)

# Print the final result
if occurrences >= 0:
    print(f"\n‚úÖ The word '{search_term}' appears")
```

```python
# Write a Python program that writes to a log file when an error occurs during file handling.
import os
import datetime

LOG_FILE = "error.log"

def log_error(error_message: str, operation: str, filename: str) -> None:
    
    # Get the current timestamp
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # Construct the log entry
    log_entry = (
        f"[{timestamp}] ERROR during '{operation}' on '{filename}'.\n"
        f"  Details: {error_message}\n"
        f"  {'='*50}\n"
    )

    try:
        # Open the log file in append mode ('a') and write the entry
        with open(LOG_FILE, 'a') as f:
            f.write(log_entry)
        print(f"--> Error logged successfully to {LOG_FILE}")
    except Exception as e:
        # Fallback error handling if the log file itself cannot be written
        print(f"FATAL: Could not write to log file {LOG_FILE}. Error: {e}")

def safe_read_file(filename: str) -> str | None:
    operation = "read"
    print(f"\n--- Attempting to {operation} file: {filename} ---")
    try:
        # 1. Check if the file is empty first (optional but good practice)
        if os.path.exists(filename) and os.path.getsize(filename) == 0:
            print("File is empty. Returning empty string.")
            return ""

        # 2. Attempt to open and read the file
        with open(filename, 'r', encoding='utf-8') as f:
            content = f.read()
            print("File read successfully.")
            return content

    except FileNotFoundError:
        # Specific exception for when the file doesn't exist
        error_msg = "File not found. Please ensure the path is correct."
        print(f"Operation failed: {error_msg}")
        log_error(error_msg, operation, filename)
        return None

    except PermissionError:
        # Specific exception for permission issues
        error_msg = "Permission denied. Check file access rights."
        print(f"Operation failed: {error_msg}")
        log_error(error_msg, operation, filename)
        return None

    except Exception as e:
        # Catch any other general file I/O errors (e.g., encoding issues)
        error_msg = str(e)
        print(f"Operation failed: An unexpected error occurred: {error_msg}")
        log_error(error_msg, operation, filename)
        return None

def main():
    # --- SCENARIO 1: Successful operation ---
    # Create a dummy file for a successful test
    existing_file = "test_data.txt"
    with open(existing_file, 'w') as f:
        f.write("This is line 1.\n")
        f.write("This is line 2.\n")

    content = safe_read_file(existing_file)
    if content is not None:
        print(f"Content read:\n---\n{content.strip()}\n---")
    
    # Clean up the dummy file
    os.remove(existing_file)


    # --- SCENARIO 2: Error simulation (FileNotFoundError) ---
    non_existent_file = "non_existent_data.dat"
    safe_read_file(non_existent_file)
    
    # --- SCENARIO 3: Error simulation (Empty file handling) ---
    empty_file = "empty.txt"
    with open(empty_file, 'w') as f:
        pass # Create an empty file
        
    safe_read_file(empty_file)
    os.remove(empty_file)

    print(f"\nCompleted execution. Check the '{LOG_FILE}' file for error details from Scenario 2.")

if __name__ == "__main__":
    main()
```
