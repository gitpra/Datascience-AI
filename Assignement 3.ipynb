```python
#Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_generator(base,exponent):
    result=1
    for i in range(exponent +1):
        yield result
        result*=base
base=int(input("enter your base number:   "))
exponent=int(input("enter your exponent number:   "))

power_generator=power_generator(base,exponent)

print(f"power of base  {base} and up to exponenet  {exponent}")
for power in power_generator:
    print(power)
```

```python
# Implement a generator function that reads a file line by line and yields each line as a string.
def file_line_reader(filepath):
    try:
        with open(filepath, 'r') as file:
            for line in file:
                yield line.strip('\n') # Strip newline characters for cleaner output
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
if __name__ == "__main__":
    # Create a dummy file for testing
    with open("sample.txt", "w") as f:
        f.write("This is line 1.\n")
        f.write("This is line 2.\n")
        f.write("And this is line 3.")
    # Iterate over the lines using the generator
    print("Reading 'sample.txt':")
    for line in file_line_reader("sample.txt"):
        print(line)

         
    
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(input_string):
    vowels="aeiouAEIOU"
    filter_chars=filter(lambda char:char not in vowels,input_string)
    return"".join(filter_chars)
my_string="Hello World"
result=remove_vowels (my_string)
print(f"original string:{my_string}")
print(f"string with vowels removed :{result}")
```
